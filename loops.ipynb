 Break and Continue

 break : used to terminate the loop when encountered.

 continue : terminate execution in the current iteration and continue execution of the loop with next iteration.


```python
i = 1
while i <= 5:
  print(i)
  if(i == 3):
       break
  i += 1
print("end of loop")
```

```python
i = 0
while i <= 5:
  i += 1
  if i == 3:
    continue
  print(i)
print(i)
```
