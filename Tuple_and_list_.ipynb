```python
#List Slicing
list = [34, 45, 46, 47, 48]
list.append(34)
print(list)

#SORTing
list = [34, 46, 48, 47, 48]
list.sort()
print(list)

#descending
list = [3, 1, 4]
print(list.sort(reverse=True))
print(list)

list = ["nuppu", "guddu", "krishu"]
print(list.sort())
print(list)

#Reversing
list = [34, 45, 46, 47, 48]
list.reverse()
print(list)

#Append
list = [34, 45, 46, 47, 48]
list.insert(0,66)
print(list)

#REMOVING
list = [34, 45, 46, 47, 48]
list.remove(34)
print(list)

list = [34, 45, 46, 47, 48]
list.pop(1)
print(list)
```

```python
#WAP to check if a list contains a palindrome of elements (hint:nuse copy())
#palindrome (racecar) starting to last same pronunce

list1 = [1, 2, 3, 1]
list2 = [1, 2, 3, 4]
copy_list1 = list1.copy()
copy_list1.reverse()
print(copy_list1)

if(list1 == copy_list1):
    print("list1 is palindrome")
else:
    print("list1 is not palindrome")











```
