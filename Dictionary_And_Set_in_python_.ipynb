Sets in python
Set is collection of unordered items

each element in the set is unique And immutable

#repeated elements stored only once, so it resolved to (1,2)

null_set = set() #empty set syntax

```python
collection = {1,2,3,4,2,"hello","world"}

print(collection)
print(type(collection))
```

```python
collection = set() #empty set;syntax
print(type(collection))
```

Set Methods

set.add(el ) #adds an element

set.remove(el ) #remove the element

set.clear() #empties the set

set.pop() #removes a random value

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add(3)
#remove
collection.remove(1)

print(collection)
```

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add("apnacollege")
collection.add((1,2,3))
collection.clear()
print(collection)
```

```python
collection = set()
collection.add(1)
collection.add(2)
collection.add("apnacollege")
collection.add((1,2,3))
collection.pop()
print(collection)
```

set.union(set2) #combines both set values and return new

set.intersection(set2) #combines common values and return new

```python
set1 = {1, 2, 3}
set2 = {2, 3, 4 }

print(set1.union(set2)) #{1,2,3,4}
```

```python
set1 = {1, 2, 3, 4}
set2 = {2, 3, 4, 5}
print(set.intersection(set1, set2))
```
